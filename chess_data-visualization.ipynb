```python
# import all essential python libraries and dependencies
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

---
## Reading Data from CSV File

```python
# Extract data from csv file
data = pd.read_csv('games.csv')
data
```

---
## Cleaning the Data

```python
# Remove rows with empty cells
data = data.dropna()

# Removes duplicate data (if any)
data.drop_duplicates(inplace=True)

data
```

```python
# Remove columns with irrelevant data
irrelevant = ['id', 'rated', 'created_at', 'last_move_at', 'increment_code', 'white_id', 'black_id']
data = data.drop(irrelevant, axis=1)
data.head()
```

```python
data.info()
```

### Wrong Data

Before we can say we have truly cleaned up the data, we are faced with one issue: wrong data. There are 4 possible victory_status which are mate, outoftime, resign, and draw. Several problems can arise from this. There can be legitimate reasons for why a player might resign or run out of time (such as when recognizing an inevitable loss or bad time management) but there are also many invalid reasons for why a player might do so (such as having to leave midgame, disconnecting, or even the other player not playing the moves they want). With these much inconsistencies, it would not produce accurate data. For example, there are many games that ended in one move because the other player ran out of time by immediately disconnecting or resigning immediately.

As such, the data has been filtered to only analyze games that ended through a draw or a mate. Even though draws can still produce inaccurate data (such as two players deciding not to play a game), it is a sacrifice that has to be made.

```python
# To eliminate any inaccuracies, only games where a mate or a draw has been delievered will be analyzed
data = data[(data['victory_status'] == 'mate') | (data['victory_status'] == 'draw')]
data.head()
```

## Data Visualization

```python
plt.figure(figsize=(9,6))
plt.title("Correlation Heatmap", fontsize=16)
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
sns.set_style("darkgrid")

plt.figure(figsize=(8,8))
plt.title("Player Rating and Outcome", fontsize=16)
plt.grid()
ax = sns.scatterplot(x="white_rating", y="black_rating", data = data, hue="winner", s = 30)
sns.regplot(data=data, x="white_rating", y="black_rating", ax = ax, scatter=False)
plt.xlabel("White Rating", fontsize=12)
plt.ylabel("Black Rating", fontsize=12)
plt.show()
```

```python
colors = ["#F9F6EE", # white
         "#808080", # draw
         "#36454F", # black
        ]

# Initialize subplot
fig, axes = plt.subplots(1, 2, figsize=(9,6))
fig.suptitle("Outcomes", fontsize=16)

# Bar Graph
ax = sns.countplot(x="winner", data=data, palette=colors, ax=axes[0])
ax = ax.bar_label(ax.containers[0])

# Pie Chart
black = data.winner.value_counts().black
draw = data.winner.value_counts().draw
white = data.winner.value_counts().white
arr = np.array([white, draw, black])
labels = ["White", "Draw", "Black"]
plt.pie(arr, labels=labels, colors=colors, autopct='%1.1f%%')

plt.show()
```
